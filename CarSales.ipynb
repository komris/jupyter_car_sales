```python

import warnings
warnings.filterwarnings('ignore')

import colorama
from colorama import Fore

import pandas as pd
from sklearn import metrics, model_selection
from sklearn.tree import DecisionTreeClassifier
import joblib
from matplotlib import pyplot
from pandas.plotting import scatter_matrix

car_data = pd.read_csv('car_data.csv')
X = car_data.drop(columns=['Purchased'])
y = car_data['Purchased']



X_train, X_test, y_train, y_test = model_selection.train_test_split(X,y,test_size=0.25)

print("Descriptive Methods")
car_data.hist()

scatter_matrix(car_data)

car_data.groupby(['Age']).sum().plot(kind='pie', y='Purchased')
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Monitoring and Maintenance")
print("Accuracy: " + str(accuracy))
if (accuracy < 0.75):
    print(Fore.RED + 'Please Inform Manager of Accuracy Number!')
else:
    print(Fore.GREEN + "Accuracy within acceptable range!")
```

```python
input("Interactive Car Sales Predictor. Press Enter key to start")
while True:
    try:
        gender = int(input("Gender of potential customer(Type 0 for female or 1 for male): "))
        if (gender == 0 or gender == 1):
            break
    except ValueError:
        print("Value Error. Please Enter a 0 or 1")
        
while True:
    try:
        age = int(input("Enter an age: "))
        if (age > 17 and age < 121):
            break
        else:
            print("Enter an age between 18 and 120")
    except ValueError:
        print("Value Error. Please enter an age number")

while True:
    try:
        salary = int(input("Enter an annual salary(digits only): "))
        break
    except ValueError:
        print("Value Error. Enter a whole number with digits only.")
predictions = model.predict([ [gender, age, salary] ])
if (predictions[0] == 0):
    print(Fore.RED + "Not Likely to Purchase a Car")
else:
    print(Fore.GREEN + "Likely to Purchase a Car")
```
